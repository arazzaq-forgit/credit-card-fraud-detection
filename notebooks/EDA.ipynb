```python
# ==========================================================
# Credit Card Fraud Detection - Exploratory Data Analysis
# ==========================================================

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------
# Load Dataset
# ----------------------------
df = pd.read_csv("../data/raw/creditcard.csv")

print("="*60)
print("Credit Card Fraud Detection Dataset")
print("="*60)

# ----------------------------
# Basic Information
# ----------------------------
print("\nDataset Shape:")
print(df.shape)

print("\nFirst Five Rows:")
display(df.head())

print("\nDataset Information:")
print(df.info())

print("\nStatistical Summary:")
display(df.describe())

# ----------------------------
# Missing Values
# ----------------------------
print("\nMissing Values:")
display(df.isnull().sum())

# ----------------------------
# Duplicate Rows
# ----------------------------
print("\nDuplicate Rows:", df.duplicated().sum())

# ----------------------------
# Data Types
# ----------------------------
print("\nData Types:")
display(df.dtypes)

# ----------------------------
# Class Distribution
# ----------------------------
print("\nClass Distribution:")
display(df["Class"].value_counts())

fraud_percent = (df["Class"].sum()/len(df))*100

print(f"\nFraud Percentage: {fraud_percent:.4f}%")

plt.figure(figsize=(6,4))
df["Class"].value_counts().plot(kind="bar")
plt.title("Fraud vs Legitimate Transactions")
plt.xticks([0,1],["Legitimate","Fraud"],rotation=0)
plt.tight_layout()
plt.show()

# ----------------------------
# Transaction Amount
# ----------------------------
plt.figure(figsize=(8,4))
plt.hist(df["Amount"],bins=50)
plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# ----------------------------
# Transaction Time
# ----------------------------
plt.figure(figsize=(8,4))
plt.hist(df["Time"],bins=50)
plt.title("Transaction Time Distribution")
plt.xlabel("Time")
plt.ylabel("Transactions")
plt.tight_layout()
plt.show()

# ----------------------------
# Correlation Matrix
# ----------------------------
plt.figure(figsize=(12,10))
corr=df.corr()
plt.imshow(corr,cmap="coolwarm")
plt.colorbar()
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()

# ----------------------------
# Top Correlated Features
# ----------------------------
corr_class = corr["Class"].sort_values(ascending=False)

print("\nTop Features Correlated with Fraud:")
display(corr_class.head(10))

print("\nMost Negatively Correlated Features:")
display(corr_class.tail(10))

# ----------------------------
# Amount Distribution by Class
# ----------------------------
plt.figure(figsize=(8,4))

plt.hist(df[df["Class"]==0]["Amount"],
         bins=50,
         alpha=0.7,
         label="Legitimate")

plt.hist(df[df["Class"]==1]["Amount"],
         bins=50,
         alpha=0.7,
         label="Fraud")

plt.legend()
plt.title("Transaction Amount by Class")
plt.xlabel("Amount")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

# ----------------------------
# Summary
# ----------------------------
print("\n" + "="*60)
print("EDA SUMMARY")
print("="*60)
print(f"Total Transactions : {len(df):,}")
print(f"Legitimate         : {(df['Class']==0).sum():,}")
print(f"Fraudulent         : {(df['Class']==1).sum():,}")
print(f"Fraud Percentage   : {fraud_percent:.4f}%")
print(f"Missing Values     : {df.isnull().sum().sum()}")
print(f"Duplicate Rows     : {df.duplicated().sum()}")

print("\nKey Observations:")
print("• Dataset is highly imbalanced.")
print("• Fraud transactions are extremely rare.")
print("• No missing values are present.")
print("• Feature scaling is required.")
print("• SMOTE or class balancing techniques are beneficial.")
print("• Random Forest is a strong baseline model for this dataset.")

print("\nEDA Completed Successfully!")
```
